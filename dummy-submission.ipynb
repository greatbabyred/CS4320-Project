```python
import numpy as np 
import pandas as pd 


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

### Load data

```python
df = pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/train.csv')
df.drop(columns=['Id'], inplace=True)

display(df.head())
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

target = df['Class']
df = df.drop(columns=['Class'])

train_x, val_x, train_y, val_y = train_test_split(df, target, test_size=0.2, random_state=42)
```

```python
## Look at summary statistics
display(df.describe())
```

### Impute Missing/NAN data

The amount of missing data as seen above is very negligible, and we don't have a large amount of data so imputing is far more favorable than dropping NaN values. 

```python
from sklearn.impute import SimpleImputer

feature_list = train_x.columns.tolist()

# Handle numeric features
numeric_features = feature_list.copy()
numeric_features.remove('EJ') # EJ is the only categorical feature
imputer = SimpleImputer(strategy='mean')
imputer.fit(train_x[numeric_features])
train_x[numeric_features] = imputer.transform(train_x[numeric_features])

# Handle categorical features
categorical_features = ['EJ']
imputer = SimpleImputer(strategy='most_frequent')
imputer.fit(train_x[categorical_features])
train_x[categorical_features] = imputer.transform(train_x[categorical_features])


```

### Scale the numeric data

The data is scaled wildy, and has a wide distrobution so this will help significantly. 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_x[numeric_features])
train_x[numeric_features] = scaler.transform(train_x[numeric_features])

display(train_x.head())
```

### Find Most Common case

```python
display(train_y.value_counts())
```

### Export a case where the most common is selected

Negative is more common so we'll generate a test submission with all negatives. They want predictions for both positive and negative seperately even though it's a binary predication, and they only give one target label. 

```python
test_df = pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/test.csv')

ids = test_df['Id']
submission = pd.DataFrame({'Id': ids, 'class_0': [1]*len(ids), 'class_1': [0]*len(ids)})
submission.to_csv('submission.csv', index=False)
```
