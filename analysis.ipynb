```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Load data

```python
df = pd.read_csv("data/train.csv")
targets = pd.read_csv("data/greeks.csv")

df = pd.merge(df, targets, on="Id")
df.drop(["Id"], axis=1, inplace=True)
display(df.head())
```

## Make Pycharts of the various target classes

```python
def make_pie_chart(data_frame, column_name):
    data_frame[column_name].value_counts().plot.pie(
        autopct="%1.1f%%",
        figsize=(10, 10),
        title=column_name,
    )
    plt.show()


make_pie_chart(df, "Alpha")
make_pie_chart(df, "Beta")
make_pie_chart(df, "Gamma")
make_pie_chart(df, "Delta")
```
