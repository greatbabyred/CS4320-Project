```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Load data

```python
df = pd.read_csv("data/train.csv")
targets = pd.read_csv("data/greeks.csv")

df = pd.merge(df, targets, on="Id")
df.drop(["Id"], axis=1, inplace=True)

target_class_list = ["Alpha", "Beta", "Gamma", "Delta"]
target_list = ["Class", "Epsilon"]
target_list.extend(target_class_list)
feature_numerical_colum_list = df.columns.drop(target_list)
feature_numerical_colum_list = feature_numerical_colum_list.drop(["EJ"])
feature_class_colum_list = ["EJ"]
```

### Make Pycharts of the various target classes

```python
def make_pie_chart(data_frame, column_name):
    data_frame[column_name].value_counts().plot.pie(
        autopct="%1.1f%%",
        figsize=(10, 10),
        title=column_name,
    )
    plt.show()



make_pie_chart(df, "Alpha")
make_pie_chart(df, "Beta")
make_pie_chart(df, "Gamma")
make_pie_chart(df, "Delta")
```

### Plot distrobition of all the numerical columns

```python
def plot_distributions(data_frame, column):
    data_frame[column].hist(bins=50, figsize=(15, 10))
    plt.xlabel(column)
    plt.title(f"Distribution of {column}")
    plt.show()


for col in feature_numerical_colum_list:
    plot_distributions(df, col)
```
